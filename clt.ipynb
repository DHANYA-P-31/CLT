```python
import numpy as np
import matplotlib.pyplot as plt

# Population parameters
mu = 0.10       # mean = 10%
sigma = 0.02    # std deviation = 2%

# Sample sizes to test
sample_sizes = [5, 20, 50, 100, 200, 500]
mean_abs_errors = []

# Run simulation
np.random.seed(42)
# for reproducibility

for n in sample_sizes:
    abs_errors = []
    for _ in range(100):  # 100 samples
        sample = np.random.normal(mu, sigma, n)
        sample_mean = np.mean(sample)
        abs_error = abs(sample_mean - mu)
        abs_errors.append(abs_error)
    mean_error = np.mean(abs_errors)
    mean_abs_errors.append(mean_error)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(sample_sizes, mean_abs_errors, color='purple', marker='o', linestyle='-')

plt.title("Mean Absolute Error vs Sample Size", fontsize=14)
plt.xlabel("Sample Size (n)", fontsize=12)
plt.ylabel("Mean Absolute Error", fontsize=12)
plt.xticks(sample_sizes)  # Show actual sample sizes on X-axis
plt.grid(True)
plt.tight_layout()
plt.show()
```

From the graph I could infers that when the sample size increases the mean of the sample is close to the population mean

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Population parameters
a, b = 100, 150
population_mean = (a + b) / 2  # 125

# Sample sizes to test
sample_sizes = [10, 20, 30, 40]
num_samples = 1000

# Store sample means for each sample size
sample_means = {n: [] for n in sample_sizes}

# Draw samples and compute means
for n in sample_sizes:
    for _ in range(num_samples):
        sample = np.random.uniform(a, b, n)
        sample_mean = np.mean(sample)
        sample_means[n].append(sample_mean)

# Calculate the mean and standard deviation of the sample means for each sample size
mean_of_sample_means = {n: np.mean(sample_means[n]) for n in sample_sizes}
std_of_sample_means = {n: np.std(sample_means[n], ddof=1) for n in sample_sizes}

# Print the mean and standard deviation of the sample means
for n in sample_sizes:
    print(f"Sample Size = {n}: Mean of sample means = {mean_of_sample_means[n]:.2f}, Standard Deviation = {std_of_sample_means[n]:.2f}")

# Plot histograms
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.ravel()

for i, n in enumerate(sample_sizes):
    axes[i].hist(sample_means[n], bins=30, density=True, alpha=0.7, color='purple')
    axes[i].set_title(f'Sample Size = {n}')
    axes[i].set_xlabel('Sample Mean')
    axes[i].set_ylabel('Density')
    # Add population mean
    axes[i].axvline(population_mean, color='red', linestyle='dashed', label='Population Mean (125)')
    # Add mean of sample means
    axes[i].axvline(mean_of_sample_means[n], color='blue', linestyle='solid', label=f'Mean of Sample Means ({mean_of_sample_means[n]:.2f})')
    # Add standard deviation to the legend
    axes[i].plot([], [], ' ', label=f'Std Dev of Sample Means ({std_of_sample_means[n]:.2f})')
    axes[i].legend()

plt.tight_layout()
# Normally, we'd use plt.show(), but we'll describe the results instead.
```
